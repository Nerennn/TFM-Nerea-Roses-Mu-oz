# **1. LIMPIEZA DE  DATOS**

Se hará:
- Eliminar las columnas que no sean "id", "timestamp", "body"
- Filtrar fechas del 28 de abril al 28 de mayo de 2025 (incluidos)
- Eliminar tweets con entradas con "RT" (eliminación de retweets)
- Añadir columna "blackout" con valores en 0 para luego codificar manualmente
- Añadir columna "nombre"

PEDRO SÁNCHEZ

```python
import pandas as pd

# Cargar el archivo CSV
file_path = r"c:\Users\nerea\Desktop\Bases de datos TFM\PEDRO SÁNCHEZ.csv"
df = pd.read_csv(file_path)

# Filtrar las columnas necesarias
df_filtered = df[["id", "timestamp", "body"]]

# Convertir la columna "timestamp" a formato datetime
df_filtered["timestamp"] = pd.to_datetime(df_filtered["timestamp"], errors="coerce")

# Filtrar las fechas entre el 28 de abril y el 28 de mayo (incluidos)
start_date = "2025-04-28"
end_date = "2025-05-29"
df_filtered = df_filtered[(df_filtered["timestamp"] >= start_date) & (df_filtered["timestamp"] <= end_date)]

# Eliminar las entradas cuyo cuerpo comience por "RT"
df_filtered = df_filtered[~df_filtered["body"].str.startswith("RT", na=False)]

# Crear la nueva columna "blackout" y rellenarla con 0
df_filtered["blackout"] = 0

# Añadir la nueva columna "nombre" al inicio del DataFrame
df_filtered.insert(0, "nombre", "Pedro Sánchez")

# Guardar el archivo con el nuevo nombre
output_path = file_path.replace(".csv", "_filtered.csv")
df_filtered.to_csv(output_path, index=False, encoding="utf-8")

print(f"Archivo filtrado guardado en: {output_path}")
```

AÑADIR UNA NUEVA COLUMNA SI/NO LÍDER EN TODAS LAS BASES 

```python
import pandas as pd
import os
import glob

def procesar_bases_datos():
    print("\n" + "="*70)
    print(" PROCESAMIENTO DE BASES DE DATOS - AÑADIR COLUMNA LÍDER ".center(70, "="))
    print("="*70 + "\n")
    
    # Directorio donde se encuentran las bases de datos
    directorio = "c:/Users/nerea/Desktop/Bases de datos TFM"
    
    # Lista de perfiles que son líderes
    lideres = ["PEDRO SÁNCHEZ", "salidayolanda", "FEIJÓO", "ABASCAL"]
    
    # Buscar todos los archivos CSV filtrados
    patron_busqueda = os.path.join(directorio, "*_filtered.csv")
    archivos = glob.glob(patron_busqueda)
    
    if not archivos:
        print(f"No se encontraron archivos con el patrón '*_filtered.csv' en {directorio}")
        return
    
    print(f"Se encontraron {len(archivos)} archivos para procesar.")
    
    # Procesar cada archivo
    for archivo in archivos:
        nombre_archivo = os.path.basename(archivo)
        nombre_base = nombre_archivo.replace("_filtered.csv", "")
        
        print(f"\nProcesando: {nombre_archivo}")
        
        # Determinar si es líder o no
        es_lider = False
        for lider in lideres:
            if lider.lower() in nombre_base.lower():
                es_lider = True
                break
        
        valor_lider = "si" if es_lider else "no"
        print(f"  Clasificado como líder: {es_lider} -> valor: '{valor_lider}'")
        
        try:
            # Cargar el dataset
            df = pd.read_csv(archivo)
            
            # Verificar si existe la columna 'body'
            if 'body' not in df.columns:
                print(f"  ¡ADVERTENCIA! El archivo {nombre_archivo} no tiene columna 'body'")
                columnas_originales = df.columns.tolist()
                # Añadir la columna 'lider' al inicio
                df.insert(0, 'lider', valor_lider)
            else:
                # Obtener el índice de la columna 'body'
                indice_body = df.columns.get_loc('body')
                
                # Añadir la columna 'lider' justo antes de 'body'
                df.insert(indice_body, 'lider', valor_lider)
            
            # Nombre del nuevo archivo
            nuevo_archivo = os.path.join(directorio, nombre_base + "_filtered_lider.csv")
            
            # Guardar el dataframe modificado
            df.to_csv(nuevo_archivo, index=False)
            
            print(f"  Guardado como: {nombre_base}_filtered_lider.csv")
            print(f"  Dimensiones: {df.shape[0]} filas × {df.shape[1]} columnas")
            
        except Exception as e:
            print(f"  ERROR al procesar {nombre_archivo}: {e}")
    
    print("\nProcesamiento completado.")

# Ejecutar la función
if __name__ == "__main__":
    procesar_bases_datos()
```

UNIÓN DE TODAS LAS BASES EN UN ARCHIVO CONJUNTO

```python
import pandas as pd
import os
import glob

def unir_bases_datos():
    print("\n" + "="*70)
    print(" UNIÓN DE BASES DE DATOS EN ARCHIVO CONJUNTO ".center(70, "="))
    print("="*70 + "\n")
    
    # Directorio donde se encuentran las bases de datos
    directorio = "c:/Users/nerea/Desktop/Bases de datos TFM"
    
    # Buscar todos los archivos CSV con el patrón "*_filtered_lider.csv"
    patron_busqueda = os.path.join(directorio, "*_filtered_lider.csv")
    archivos = glob.glob(patron_busqueda)
    
    if not archivos:
        print(f"No se encontraron archivos con el patrón '*_filtered_lider.csv' en {directorio}")
        return
    
    print(f"Se encontraron {len(archivos)} archivos para unir:")
    for idx, archivo in enumerate(archivos, 1):
        print(f"  {idx}. {os.path.basename(archivo)}")
    
    # Lista para almacenar los dataframes
    dfs = []
    total_filas = 0
    
    # Procesar cada archivo
    for archivo in archivos:
        nombre_archivo = os.path.basename(archivo)
        nombre_base = nombre_archivo.replace("_filtered_lider.csv", "")
        
        try:
            # Cargar el dataset
            df = pd.read_csv(archivo)
            
            # Añadir columna para identificar el origen
            df['origen'] = nombre_base
            
            # Almacenar el dataframe
            dfs.append(df)
            
            # Actualizar contador de filas
            total_filas += len(df)
            
            print(f"Cargado: {nombre_archivo} - {len(df)} filas")
            
        except Exception as e:
            print(f"ERROR al cargar {nombre_archivo}: {e}")
    
    if not dfs:
        print("No se pudo cargar ningún archivo. Operación abortada.")
        return
    
    # Unir todos los dataframes
    print("\nUniendo dataframes...")
    df_combinado = pd.concat(dfs, ignore_index=True)
    
    # Verificar que la unión se realizó correctamente
    print(f"Dataframe combinado: {df_combinado.shape[0]} filas × {df_combinado.shape[1]} columnas")
    
    if df_combinado.shape[0] != total_filas:
        print("¡ADVERTENCIA! El número de filas no coincide con la suma esperada.")
    
    # Ruta del archivo de salida
    archivo_salida = os.path.join(directorio, "base_conjunta.csv")
    
    # Guardar el dataframe combinado
    df_combinado.to_csv(archivo_salida, index=False)
    
    print(f"\nBase de datos conjunta guardada como: {archivo_salida}")
    print(f"La base conjunta contiene datos de {len(dfs)} archivos originales.")
    
    # Mostrar información de las columnas
    print("\nColumnas en la base conjunta:")
    for col in df_combinado.columns:
        print(f"  - {col}")
    
    # Mostrar distribución de orígenes
    print("\nDistribución por origen:")
    distribucion = df_combinado['origen'].value_counts().sort_values(ascending=False)
    for origen, count in distribucion.items():
        print(f"  - {origen}: {count} filas")
    
    print("\nProceso completado con éxito.")

# Ejecutar la función
if __name__ == "__main__":
    unir_bases_datos()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
from collections import Counter
import re
import nltk
from nltk.corpus import stopwords

# Configurar pandas para mostrar todas las filas sin truncamiento
pd.set_option('display.max_rows', None)
pd.set_option('display.max_columns', None)
pd.set_option('display.width', 1000)
pd.set_option('display.expand_frame_repr', False)

# Crear una lista con todos los archivos CSV filtrados
archivos = [
    "PEDRO SÁNCHEZ_filtered.csv",
    "ABASCAL_filtered.csv",
    "FEIJÓO_filtered.csv",
    "PP_filtered.csv",
    "PSOE_filtered.csv",
    "VOX_filtered.csv",
    "salidasumar_filtered.csv",
    "salidayolanda_filtered.csv"
]

# Función para obtener la columna de fecha del dataset
def get_date_column(df):
    if 'timestamp' in df.columns:
        return 'timestamp'
    elif 'created' in df.columns:
        return 'created'
    elif 'date' in df.columns:
        return 'date'
    return None

# Función para obtener la columna de contenido del dataset
def get_content_column(df):
    if 'body' in df.columns:
        return 'body'
    elif 'text' in df.columns:
        return 'text'
    return None

# Cargar todos los datasets en un diccionario
datasets = {}
for archivo in archivos:
    nombre = archivo.replace("_filtered.csv", "")
    ruta = f"c:/Users/nerea/Desktop/Bases de datos TFM/{archivo}"
    try:
        datasets[nombre] = pd.read_csv(ruta)
        print(f"Cargado {nombre}: {len(datasets[nombre])} tweets")
        print(f"  Columnas disponibles: {', '.join(datasets[nombre].columns.tolist())}")
    except Exception as e:
        print(f"Error al cargar {nombre}: {e}")

# Crear un archivo de texto para guardar los resultados completos
with open('estadisticas_exploratorias.txt', 'w', encoding='utf-8') as f:
    # Ver estadísticas básicas para cada dataset
    for nombre, df in datasets.items():
        info = f"\n{'='*50}\n"
        info += f"ESTADÍSTICAS PARA: {nombre}\n"
        info += f"{'='*50}\n"
        info += f"Número total de tweets: {len(df)}\n"
        
        # Obtener la columna de fecha adecuada
        date_col = get_date_column(df)
        if date_col:
            info += f"Período de tiempo: {df[date_col].min()} a {df[date_col].max()}\n"
        else:
            info += "No se encontró columna de fecha/hora\n"
        
        # Obtener la columna de contenido adecuada
        content_col = get_content_column(df)
        if content_col:
            info += f"Longitud media de tweets: {df[content_col].str.len().mean():.2f} caracteres\n"
            info += f"Tweet más corto: {df[content_col].str.len().min()} caracteres\n"
            info += f"Tweet más largo: {df[content_col].str.len().max()} caracteres\n"
        else:
            info += "No se encontró columna de contenido de tweet\n"
            
        # Escribir en el archivo y mostrar en pantalla
        f.write(info)
        print(info)

print(f"\nEstadísticas guardadas en el archivo: estadisticas_exploratorias.txt")
```

# **3. ANÁLISIS TEMPORAL**

```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
import seaborn as sns
from datetime import datetime
import numpy as np
import os
import time

# Generar un timestamp único para los archivos
timestamp = int(time.time())

# Configuración estética para gráficos
plt.style.use('default')
sns.set_theme(style="whitegrid")
plt.rcParams.update({'font.size': 11, 'figure.figsize': (12, 7)})

# Definir esquema de colores político
party_colors = {
    # PSOE y Pedro Sánchez (rojos)
    "PSOE": "darkred",
    "Pedro Sánchez": "#FF6666",
    "PEDRO SÁNCHEZ": "#FF6666",
    
    # PP y Feijóo (azules)
    "PP": "darkblue",
    "Feijóo": "#6666FF",
    "FEIJÓO": "#6666FF",
    
    # VOX y Abascal (verdes)
    "VOX": "darkgreen",
    "ABASCAL": "#66FF66",
    
    # Sumar y Yolanda Díaz (morados)
    "Sumar": "darkviolet",
    "SUMAR": "darkviolet",
    "Yolanda Díaz": "#CC99FF",
    "YOLANDA DÍAZ": "#CC99FF"
}

# SOLUCIÓN DEFINITIVA: Diccionario para renombrar perfiles
RENOMBRAR_PERFILES = {
    "salidasumar": "SUMAR",
    "salidayolanda": "YOLANDA DÍAZ"
}

# Función para obtener el color político correcto
def get_political_color(nombre):
    # Si el nombre es uno de los que vamos a renombrar, usar el nombre nuevo para buscar color
    if nombre in RENOMBRAR_PERFILES:
        nombre_buscar = RENOMBRAR_PERFILES[nombre]
        if nombre_buscar in party_colors:
            return party_colors[nombre_buscar]
    
    # Intentar obtener el color directo
    if nombre in party_colors:
        return party_colors[nombre]
    
    # Si no hay coincidencia exacta, buscar coincidencia parcial
    nombre_lower = nombre.lower()
    
    if "sanchez" in nombre_lower or "sánchez" in nombre_lower:
        return party_colors["Pedro Sánchez"]
    elif "feij" in nombre_lower:
        return party_colors["Feijóo"]
    elif "abas" in nombre_lower:
        return party_colors["ABASCAL"]
    elif "vox" in nombre_lower:
        return party_colors["VOX"]
    elif "psoe" in nombre_lower:
        return party_colors["PSOE"]
    elif "pp" in nombre_lower:
        return party_colors["PP"]
    elif "sumar" in nombre_lower:
        return party_colors["SUMAR"]
    elif "yolanda" in nombre_lower or "diaz" in nombre_lower or "díaz" in nombre_lower:
        return party_colors["YOLANDA DÍAZ"]
    
    # Si no hay coincidencia, usar gris
    return "gray"

# Cargar todos los datasets en un diccionario
def cargar_datasets():
    # Crear una lista con todos los archivos CSV filtrados
    archivos = [
        "PEDRO SÁNCHEZ_filtered.csv",
        "ABASCAL_filtered.csv",
        "FEIJÓO_filtered.csv",
        "PP_filtered.csv",
        "PSOE_filtered.csv",
        "VOX_filtered.csv",
        "salidasumar_filtered.csv",
        "salidayolanda_filtered.csv"
    ]
    
    datasets = {}
    for archivo in archivos:
        nombre = archivo.replace("_filtered.csv", "")
        ruta = f"c:/Users/nerea/Desktop/Bases de datos TFM/{archivo}"
        try:
            df = pd.read_csv(ruta)
            
            # Normalizar el formato de fecha
            if 'timestamp' in df.columns:
                df['timestamp'] = pd.to_datetime(df['timestamp'], errors='coerce')
                if hasattr(df['timestamp'].dt, 'tz_localize'):
                    df['timestamp'] = df['timestamp'].dt.tz_localize(None)
            
            # Guardar en el diccionario
            datasets[nombre] = df
            print(f"Cargado {nombre}: {len(df)} tweets")
            
        except Exception as e:
            print(f"Error al cargar {nombre}: {e}")
    
    return datasets

# Función para preparar los datos temporales
def prepare_temporal_data(datasets):
    processed_dfs = {}
    
    for nombre, df in datasets.items():
        if len(df) == 0:
            continue
            
        date_col = None
        if 'timestamp' in df.columns:
            date_col = 'timestamp'
        elif 'created' in df.columns:
            date_col = 'created'
        
        if date_col is None or 'blackout' not in df.columns:
            continue
            
        temp_df = df.copy()
        temp_df['fecha'] = temp_df[date_col].dt.date
        temp_df['blackout'] = temp_df['blackout'].astype(int)
        
        # IMPORTANTE: Renombrar el perfil si está en el diccionario
        if nombre in RENOMBRAR_PERFILES:
            nuevo_nombre = RENOMBRAR_PERFILES[nombre]
            print(f"RENOMBRANDO: {nombre} -> {nuevo_nombre}")
            processed_dfs[nuevo_nombre] = temp_df
        else:
            processed_dfs[nombre] = temp_df
    
    return processed_dfs

# Función para crear un gráfico comparativo de todos los perfiles
def plot_combined_blackout_evolution(processed_dfs):
    # Imprimir los perfiles después del renombrado
    print("\nPerfiles después del renombrado:")
    for nombre in processed_dfs.keys():
        print(f"  - {nombre}")
    
    # Crear un dataframe combinado
    combined_data = pd.DataFrame()
    
    # Determinar el rango de fechas global
    min_date = pd.Timestamp('2025-04-28').date()
    max_date = pd.Timestamp('2025-05-29').date()
    
    # Procesar cada perfil
    for nombre, df in processed_dfs.items():
        # Crear rango completo de fechas
        date_range = pd.date_range(start=min_date, end=max_date).date
        
        # Crear dataframe con fechas completas
        full_dates = pd.DataFrame({
            'fecha': np.repeat(date_range, 2),
            'blackout': np.tile([0, 1], len(date_range))
        })
        
        # Contar tweets por fecha y blackout
        if 'fecha' in df.columns and 'blackout' in df.columns:
            df_temp = df.copy()
            df_temp['nombre'] = nombre  # Usar el nombre ya renombrado
            actual_counts = df_temp.groupby(['fecha', 'blackout']).size().reset_index(name='count')
            actual_counts['nombre'] = nombre
            
            merged_counts = pd.merge(full_dates, actual_counts, 
                                    on=['fecha', 'blackout'], 
                                    how='left').fillna(0)
            merged_counts['nombre'] = nombre
        else:
            merged_counts = full_dates.copy()
            merged_counts['count'] = 0
            merged_counts['nombre'] = nombre
        
        combined_data = pd.concat([combined_data, merged_counts])
    
    # Convertir fecha a datetime
    combined_data['fecha'] = pd.to_datetime(combined_data['fecha'])
    
    # Gráfico 1: Tweets regulares (blackout=0)
    plt.figure(figsize=(14, 8))
    
    for nombre in sorted(processed_dfs.keys()):
        data_perfil = combined_data[(combined_data['nombre'] == nombre) & 
                                   (combined_data['blackout'] == 0)]
        
        if len(data_perfil) > 0:
            data_perfil = data_perfil.sort_values('fecha')
            plt.plot(data_perfil['fecha'], data_perfil['count'], 
                     label=nombre,  # Nombre ya renombrado
                     marker='o', linestyle='-', 
                     color=get_political_color(nombre))
            print(f"Graficado perfil '{nombre}' (tweets regulares)")
    
    plt.gca().xaxis.set_major_formatter(mdates.DateFormatter('%d-%b'))
    plt.gca().xaxis.set_major_locator(mdates.DayLocator(interval=2))
    plt.title('Evolución temporal de tweets regulares por cuenta', fontsize=14)
    plt.xlabel('Fecha', fontsize=12)
    plt.ylabel('Número de tweets', fontsize=12)
    # Leyenda dentro del gráfico para tweets regulares
    plt.legend(title='Cuenta', loc='upper right', framealpha=0.7, fontsize=9)
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.grid(True, alpha=0.3)
    
    # Guardar con timestamp para evitar problemas de caché
    filename = f'evolucion_tweets_regulares_{timestamp}.png'
    plt.savefig(filename, dpi=300)
    print(f"Guardado gráfico como: {filename}")
    plt.close()
    
    # Gráfico 2: Tweets sobre apagón (blackout=1)
    plt.figure(figsize=(14, 8))
    
    for nombre in sorted(processed_dfs.keys()):
        data_perfil = combined_data[(combined_data['nombre'] == nombre) & 
                                   (combined_data['blackout'] == 1)]
        
        if len(data_perfil) > 0:
            data_perfil = data_perfil.sort_values('fecha')
            plt.plot(data_perfil['fecha'], data_perfil['count'], 
                     label=nombre,  # Nombre ya renombrado
                     marker='o', linestyle='-', 
                     color=get_political_color(nombre))
            print(f"Graficado perfil '{nombre}' (tweets apagón)")
    
    plt.gca().xaxis.set_major_formatter(mdates.DateFormatter('%d-%b'))
    plt.gca().xaxis.set_major_locator(mdates.DayLocator(interval=2))
    plt.title('Evolución temporal de tweets sobre el apagón por cuenta', fontsize=14)
    plt.xlabel('Fecha', fontsize=12)
    plt.ylabel('Número de tweets', fontsize=12)
    
    # Leyenda dentro del gráfico para tweets del apagón
    plt.legend(title='Cuenta', loc='upper right', framealpha=0.7, fontsize=9)
    
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.grid(True, alpha=0.3)
    
    # Guardar con timestamp para evitar problemas de caché
    filename = f'evolucion_tweets_apagon_{timestamp}.png'
    plt.savefig(filename, dpi=300)
    print(f"Guardado gráfico como: {filename}")
    plt.close()
    
    # Mostrar directorio donde se guardaron las imágenes
    print(f"\nLos gráficos se han guardado en: {os.getcwd()}")

# Función principal
def analisis_temporal_blackout():
    print("\n" + "="*70)
    print(" ANÁLISIS TEMPORAL DE TWEETS (NOMBRES CORREGIDOS) ".center(70, "="))
    print("="*70 + "\n")
    
    # Cargar datos
    datasets = cargar_datasets()
    
    # Preparar los datos (ya con nombres corregidos)
    processed_dfs = prepare_temporal_data(datasets)
    
    # Generar gráficos comparativos
    plot_combined_blackout_evolution(processed_dfs)
    
    print("\nAnálisis temporal completado con nombres corregidos en la leyenda.")

# Ejecutar el análisis
analisis_temporal_blackout()
```

# **4. ANÁLISIS DE TEMAS (LDA)**

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import re
import os
import seaborn as sns
from collections import Counter
import nltk
from nltk.corpus import stopwords
import warnings
from sklearn.decomposition import LatentDirichletAllocation
from sklearn.feature_extraction.text import CountVectorizer

# Configuraciones básicas
warnings.filterwarnings("ignore")
plt.style.use('seaborn-v0_8-whitegrid')
sns.set(font_scale=1.1)
nltk.download('stopwords', quiet=True)

# Directorio para resultados
RESULTADOS_DIR = 'resultados_lda_completo'
os.makedirs(RESULTADOS_DIR, exist_ok=True)

# Función para cargar datasets
def cargar_datasets():
    archivos = [
        "PEDRO SÁNCHEZ_filtered.csv",
        "ABASCAL_filtered.csv",
        "FEIJÓO_filtered.csv",
        "PP_filtered.csv",
        "PSOE_filtered.csv",
        "VOX_filtered.csv",
        "salidasumar_filtered.csv",
        "salidayolanda_filtered.csv"
    ]
    
    datasets = {}
    for archivo in archivos:
        nombre = archivo.replace("_filtered.csv", "")
        ruta = f"c:/Users/nerea/Desktop/Bases de datos TFM/{archivo}"
        try:
            datasets[nombre] = pd.read_csv(ruta)
            print(f"Cargado {nombre}: {len(datasets[nombre])} tweets")
        except Exception as e:
            print(f"Error al cargar {nombre}: {e}")
    
    return datasets

# Función para preprocesar texto
def preprocesar_texto(texto):
    if pd.isna(texto) or not isinstance(texto, str):
        return ""
    
    texto = texto.lower()
    texto = re.sub(r'https?://\S+|www\.\S+', '', texto)
    texto = re.sub(r'@\S+', '', texto)
    texto = re.sub(r'#(\w+)', r'\1', texto)
    texto = re.sub(r'[^\w\s]', '', texto)
    texto = re.sub(r'\d+', '', texto)
    texto = re.sub(r'\s+', ' ', texto).strip()
    
    return texto

# Obtener lista de stopwords mejorada
def obtener_stopwords():
    try:
        stop_words = set(stopwords.words('spanish'))
    except:
        stop_words = set()
    
    # Añadir stopwords adicionales comunes
    stop_words.update([
        'rt', 'xd', 'jaja', 'jajaja', 'si', 'no', 'q', 'd', 'k', 'x', 
        'el', 'la', 'los', 'las', 'un', 'una', 'y', 'o', 'pero', 'porque', 
        'como', 'cuando', 'donde', 'quien', 'que', 'este', 'esta', 'estos', 
        'estas', 'ese', 'esa', 'esos', 'esas', 'del', 'al', 'lo', 'le', 'les', 
        'me', 'te', 'se', 'nos', 'para', 'por', 'con', 'sin', 'sobre', 'entre', 
        'hasta', 'desde', 'contra', 'durante', 'mediante', 'según', 'a', 'ante', 
        'bajo', 'cabe', 'de', 'en', 'nuevo', 'gran', 'partido', 'vamos', 'país',
        'va', 'ser', 'hace', 'hacer', 'han', 'hay', 'hoy', 'así', 'cada', 'todo', "mayo", 
        "día", "consigue", "llegar", "días", "agenda", "buenos", "vuelta", "puede", 
        "pierdas", "partir", "menos", "puede", "zona", "roja", "vez", "siempre", "portavoz", 
        "sede", "aquí", "españa", "hacia", "años", "ahora", "todas", "europa",
        "mejor", "usted", "horas", "ver", "buenas", "compartiremos", "peor", "mientras", 
        "frente", "gracias", "ejecutiva", "josé", "compartimos", "españoles", "causa",
        "comisión", "gente"
    ])
    
    return stop_words

# Función para analizar temas con LDA y generar resultados
def analizar_temas(datasets, num_temas=5):
    # Obtener stopwords
    stop_words = obtener_stopwords()
    
    # Diccionario para almacenar resultados
    resultados = {}
    tema_principal_por_perfil = {}
    
    # Analizar cada perfil
    for nombre, df in datasets.items():
        print(f"\n{'='*50}")
        print(f"Analizando temas para: {nombre}")
        
        # Usar todos los datos
        df_completo = df.copy()
        
        if len(df_completo) < 5:
            print(f"Insuficientes tweets para {nombre}")
            continue
            
        # Preprocesar textos
        if 'body' in df_completo.columns:
            textos = df_completo['body'].fillna('').apply(preprocesar_texto).tolist()
            textos = [t for t in textos if len(t) > 10]  # Filtrar textos muy cortos
            
            if len(textos) < 5:
                print(f"Insuficientes textos válidos para {nombre}")
                continue
            
            # Vectorizar textos
            vectorizer = CountVectorizer(
                max_df=0.95,
                min_df=2,
                stop_words=list(stop_words),
                max_features=10000
            )
            
            X = vectorizer.fit_transform(textos)
            feature_names = vectorizer.get_feature_names_out()
            
            # Crear y entrenar modelo LDA
            lda = LatentDirichletAllocation(
                n_components=min(num_temas, max(2, len(textos)//50)),
                random_state=42,
                max_iter=30,
                learning_method='online',
                learning_offset=50.0,
                batch_size=128
            )
            
            lda.fit(X)
            
            # Guardar resultados generales
            resultados[nombre] = {
                'modelo': lda,
                'vectorizador': vectorizer,
                'feature_names': feature_names,
                'num_temas': lda.n_components,
                'num_tweets': len(textos)
            }
            
            # Extraer el tema principal (el primero)
            tema_principal = lda.components_[0]
            top_palabras_idx = tema_principal.argsort()[:-11:-1]  # Top 10 palabras
            top_palabras = [feature_names[i] for i in top_palabras_idx]
            
            # Guardar el tema principal para este perfil
            tema_principal_por_perfil[nombre] = top_palabras
            
            print(f"Tema principal: {', '.join(top_palabras)}")
            
            # Guardar en archivo de texto detallado
            with open(f'{RESULTADOS_DIR}/temas_{nombre.replace(" ", "_")}.txt', 'w', encoding='utf-8') as f:
                f.write(f"Análisis de temas para {nombre} (TODOS LOS TWEETS)\n")
                f.write(f"Total de tweets analizados: {len(textos)}\n\n")
                
                for tema_idx, tema in enumerate(lda.components_):
                    top_palabras_idx = tema.argsort()[:-31:-1]  # 30 palabras principales
                    top_palabras_tema = [feature_names[i] for i in top_palabras_idx]
                    f.write(f"Tema {tema_idx+1}: {', '.join(top_palabras_tema)}\n\n")
    
    # Generar tabla resumen de temas principales
    generar_tabla_temas_principales(tema_principal_por_perfil)
    
    return resultados

# Función para generar tabla de temas principales
def generar_tabla_temas_principales(tema_principal_por_perfil):
    print("\n" + "="*70)
    print(" TABLA DE TÉRMINOS PRINCIPALES POR PERFIL ".center(70, "="))
    print("="*70 + "\n")
    
    # Definir mapeo de nombres de perfiles
    nombre_display = {
        "salidasumar": "SUMAR",
        "salidayolanda": "YOLANDA DÍAZ"
    }
    
    # Definir orden de perfiles
    orden_perfiles = [
        "PSOE", 
        "PEDRO SÁNCHEZ", 
        "PP", 
        "FEIJÓO", 
        "VOX", 
        "ABASCAL", 
        "salidasumar", 
        "salidayolanda"
    ]
    
    # Crear filas para la tabla
    filas = []
    
    # Procesar perfiles en el orden deseado
    for perfil in orden_perfiles:
        # Buscar el perfil (considerando mayúsculas/minúsculas)
        perfil_encontrado = None
        for p in tema_principal_por_perfil.keys():
            if p.lower() == perfil.lower():
                perfil_encontrado = p
                break
        
        if perfil_encontrado:
            terminos = tema_principal_por_perfil[perfil_encontrado]
            
            # Determinar el nombre a mostrar
            nombre_mostrar = nombre_display.get(perfil_encontrado, perfil_encontrado)
            
            # Crear fila [Cuenta, Palabra1, Palabra2, Palabra3]
            fila = [nombre_mostrar]
            
            # Añadir los tres términos principales
            for i in range(3):
                if i < len(terminos):
                    fila.append(terminos[i])
                else:
                    fila.append("--")
            
            filas.append(fila)
    
    # Crear DataFrame
    columnas = ['Cuenta', 'Palabra 1', 'Palabra 2', 'Palabra 3']
    df_tabla = pd.DataFrame(filas, columns=columnas)
    
    # Guardar como CSV
    df_tabla.to_csv(f'{RESULTADOS_DIR}/tabla_temas_principales.csv', index=False)
    print(f"Tabla de temas principales guardada como CSV: {RESULTADOS_DIR}/tabla_temas_principales.csv")
    
    # Guardar como Excel
    df_tabla.to_excel(f'{RESULTADOS_DIR}/tabla_temas_principales.xlsx', index=False)
    print(f"Tabla de temas principales guardada como Excel: {RESULTADOS_DIR}/tabla_temas_principales.xlsx")
    
    # Mostrar tabla en consola
    print("\nTabla de términos principales por perfil:\n")
    print(df_tabla.to_string(index=False))
    
    # Crear visualización gráfica de la tabla
    plt.figure(figsize=(12, len(filas)*0.7 + 1.5))
    
    # Crear tabla como imagen
    ax = plt.subplot(111)
    ax.axis('off')
    
    # Crear tabla
    tabla = plt.table(
        cellText=df_tabla.values,
        colLabels=df_tabla.columns,
        cellLoc='center',
        loc='center',
        colWidths=[0.25, 0.25, 0.25, 0.25]
    )
    
    # Dar formato a la tabla
    tabla.auto_set_font_size(False)
    tabla.set_fontsize(11)
    tabla.scale(1, 1.5)
    
    # Colorear encabezados
    for j in range(len(columnas)):
        tabla[(0, j)].set_facecolor('#4472C4')
        tabla[(0, j)].set_text_props(color='white', fontweight='bold')
    
    # Colorear filas por familia política
    colores_politicos = {
        "PSOE": "#E61919",      # Rojo
        "PEDRO SÁNCHEZ": "#E61919",
        "PP": "#0055A4",        # Azul
        "FEIJÓO": "#0055A4",
        "VOX": "#5AC035",       # Verde
        "ABASCAL": "#5AC035",
        "SUMAR": "#6A1B9A",     # Morado
        "YOLANDA DÍAZ": "#6A1B9A"
    }
    
    for i, fila in enumerate(filas, 1):
        perfil = fila[0]  # Primera columna es el perfil
        
        if perfil in colores_politicos:
            color = colores_politicos[perfil]
            # Aplicar color con transparencia
            for j in range(len(columnas)):
                rgba = plt.matplotlib.colors.to_rgba(color, alpha=0.2)
                tabla[(i, j)].set_facecolor(rgba)
    
    plt.title('Términos principales por perfil político', fontsize=16)
    plt.tight_layout()
    
    # Guardar la tabla como imagen
    plt.savefig(f'{RESULTADOS_DIR}/tabla_temas_principales.png', dpi=300, bbox_inches='tight')
    plt.close()
    
    print(f"Visualización de tabla guardada como: {RESULTADOS_DIR}/tabla_temas_principales.png")

# Función principal
def analisis_lda_principal():
    print("\n" + "="*70)
    print(" ANÁLISIS LDA: TEMA PRINCIPAL POR PERFIL ".center(70, "="))
    print("="*70 + "\n")
    
    # Cargar datasets
    print("Cargando datasets...")
    datasets = cargar_datasets()
    
    if not datasets:
        print("No se pudieron cargar los datasets")
        return
    
    # Analizar temas
    print("\nExtrayendo tema principal de cada perfil político...")
    resultados = analizar_temas(datasets, num_temas=5)
    
    if resultados:
        print("\nAnálisis completado con éxito.")
        print(f"Los resultados están disponibles en el directorio '{RESULTADOS_DIR}'")
    else:
        print("\nNo se pudieron generar resultados del análisis")

# Ejecutar análisis
analisis_lda_principal()
```

# **5. MODELO DE RED NEURONAL**

```python
from transformers import AutoTokenizer, AutoModelForSequenceClassification, pipeline

# Cargar el modelo y el tokenizer de RoBERTuito entrenado para análisis de sentimiento
modelo = "pysentimiento/robertuito-sentiment-analysis"
tokenizer = AutoTokenizer.from_pretrained(modelo)
model = AutoModelForSequenceClassification.from_pretrained(modelo)

# Crear pipeline para clasificación
analizador = pipeline("sentiment-analysis", model=model, tokenizer=tokenizer)
```

EJEMPLO

```python
# Frase de ejemplo
frase1 = "Estoy harto de todo esto"
frase2 = "Me encanta la programación en Python"
frase3 = "Hoy hace sol"

# Clasificar
resultado = analizador(frase1)
print(resultado)
resultado = analizador(frase2)
print(resultado)
resultado = analizador(frase3)
print(resultado)
```

Como se puede apreciar, el modelo de red neuronal da una salida del tipo:
- (sentimiento, score)
donde sentimiento puede ser NEG (negativo), NEU (neutral) o POS (positivo) y score es un número entre 0 y 1, indicando una puntuación asignada por el modelo al sentimiento inferido.

Por tanto, no se trata de un modelo de regresión, o de clasificación binaria, sino de un modelo de clasificación multiclase, donde la salida es la tupla (sentimiento, score).

De cara a poder agregar valores de score, es necesario transformar de acuerdo a nuestro interés. No podemos simplemente sumar los scores, ya que el score por si solo no indica el sentimiento de la frase. En nuestro caso queremos diferenciar entre sentimiento (positivo o negativo) y neutralidad. Por tanto, tiene sentido utilizar el score como un peso y asignar un valor numérico a las etiquetas:

- NEG = 1
- NEU = -1
- POS = 1

Como solo queremos diferenciar entre sentimiento y no sentimiento (para la hipótesis 3), asignamos el valor 1 a NEG y POS, y -1 a NEU. De esta forma, si la frase es positiva o negativa, el score se suma al resultado final, mientras que si es neutral, se resta.

Cada frase quedaría entonces caracterizada por la tupla (sentimiento, score) y una puntuación calculada como:
$$
\text{puntuacion} = \text{sentimiento} \cdot \text{score} 
$$
siendo sentimiento el valor numérico asignado a la etiqueta (-1 o 1).

# **6. ANÁLISIS DE SENTIMIENTOS (HIPÓTESIS 2)**

```python
# Generamos un conjunto de datos de ejemplo
import pandas as pd
import numpy as np

lider_a_partido={}
data = pd.read_csv("base_conjunta.csv")
# data["Partido"]
```

```python
lider_a_partido ={'Santiago Abascal':"VOX", 'Feijóo':"PP", 'Pedro Sánchez':"PSOE", 'PP':"PP", 'PSOE':"PSOE",
       'Sumar':"Sumar", 'Yolanda Díaz':"Sumar", 'VOX':'VOX'}
oposicion = set(['PP', 'VOX', 'Sumar'])
# Añadimos el nombre del partido a cada líder
data["Partido"] = data["nombre"].map(lider_a_partido)
# Añadimos si se trata de oposición
data["Oposicion"] = data["Partido"].apply(lambda x: x in oposicion)
# Filtramos para que solo sea del blackout
data = data[data["blackout"]==1]
```

```python
data.head()
```

```python
from tqdm import tqdm
sentimiento = []
score = []
puntuacion = []
dic = {"NEU":-1, "POS":1, "NEG":1}

for frase in tqdm(data["body"]):
  try:
    resultado = analizador(frase)
  except Exception as e:
    print(f"Error procesando la frase: {frase}")
    print(f"Error: {e}")
  # Guardamos la etiqueta de sentimiento
  sentimiento.append(resultado[0]['label'])
  # guardamos el score
  score.append(resultado[0]['score'])
  
  # calculamos la puntuación multiplicando el score por el valor de la etiqueta
  puntuacion.append(dic[resultado[0]['label']]*resultado[0]['score'])

  
data["Sentimiento"] = sentimiento
data["Score"] = score
data["Puntuacion"] = puntuacion
df = pd.DataFrame(data)
```

```python
print("Resultados")
print(df)
print("\n")

# Sumamos puntuacion agrupado por partido
print("Resultados por partido (sumando la puntuacion)")
df_partido = df.groupby("Partido")["Puntuacion"].sum().reset_index()
print(df_partido)
print("\n")
# Sumamos puntuacion agrupado por partido y lider
print("Resultados por partido y lider (sumando la puntuacion)")
df_partido_lider = df.groupby(["Partido", "Lider"])["Puntuacion"].sum().reset_index()
print(df_partido_lider)
print("\n")
```

```python
# Confianza de la oposicion que sea negativo
# Tweets negativos de la oposicion
df_oposicion = df[df["Oposicion"] == True]
df_oposicion_neg = df_oposicion[df_oposicion["Sentimiento"] == "NEG"]


confianza = df_oposicion_neg.shape[0] / df_oposicion.shape[0]
print("Confianza de la oposicion que sea negativo: ", confianza)

# Lift
lift = confianza / (df[df["Sentimiento"] == "NEG"].shape[0] / df.shape[0])
print("Tweets negativos de la oposicion")
print(df_oposicion_neg.shape[0])
print("Tweets de la oposicion")
print(df_oposicion.shape[0])
print("Total de tweets negativos")
print(df[df["Sentimiento"] == "NEG"].shape[0])
print("Lift: ", lift)
```

# **7. COMPROBACIÓN HIPÓTESIS 3**

```python

data_ps = data[data["nombre"]=="Pedro Sánchez"]
data_nf = data[data["nombre"]=="Feijóo"]
data_ab = data[data["nombre"]=="Santiago Abascal"]
data_yd = data[data["nombre"]=="Yolanda Díaz"]
data_su = data[data["nombre"]=="Sumar"]
data_pp = data[data["nombre"]=="PP"]
data_vox = data[data["nombre"]=="VOX"]
data_psoe = data[data["nombre"]=="PSOE"]
```

HIPÓTESIS DE NORMALIDAD

```python
from scipy.stats import kstest, ttest_ind, mannwhitneyu

# Aplicar test de normalidad a cada dataset usando {}

# Aplicar test de hipótesis por pares (en Puntuación)
# Usando el modelo {modelo}


# Comprobar normalidad con Kolmogórov-Smirnov
normalidad_pp = kstest(data_pp["Puntuacion"], 'norm')
normalidad_psoe = kstest(data_psoe["Puntuacion"], 'norm')
normalidad_su = kstest(data_su["Puntuacion"], 'norm')
normalidad_vox = kstest(data_vox["Puntuacion"], 'norm')

normalidad_ps = kstest(data_ps["Puntuacion"], 'norm')
normalidad_nf = kstest(data_nf["Puntuacion"], 'norm')
normalidad_ab = kstest(data_ab["Puntuacion"], 'norm')
normalidad_yd = kstest(data_yd["Puntuacion"], 'norm')

print("Normalidad PP:", normalidad_pp.pvalue)
print("Normalidad Feijóo:", normalidad_nf.pvalue)

print("Normalidad PSOE:", normalidad_psoe.pvalue)
print("Normalidad Pedro Sánchez:", normalidad_ps.pvalue)

print("Normalidad Sumar:", normalidad_su.pvalue)
print("Normalidad Yolanda Díaz:", normalidad_yd.pvalue)

print("Normalidad VOX:", normalidad_vox.pvalue)
print("Normalidad Santiago Abascal:", normalidad_ab.pvalue)
```

HIPÓTESIS DE MEDIAS

```python
# PSOE y PS son normales
t_test_result_psoe = ttest_ind(data_psoe["Puntuacion"], data_ps["Puntuacion"])
print("Medias de puntuación")
print("PSOE:", data_psoe["Puntuacion"].mean())
print("Pedro Sánchez:", data_ps["Puntuacion"].mean())
print("Prueba t de Student entre PSOE y Pedro Sánchez:", t_test_result_psoe.pvalue)
print("\n")

# Sumar y YD son normales
t_test_result_sumar = ttest_ind(data_su["Puntuacion"], data_yd["Puntuacion"])
print("Medias de puntuación")
print("Sumar:", data_su["Puntuacion"].mean())
print("Yolanda Díaz:", data_yd["Puntuacion"].mean())
print("Prueba t de Student entre Sumar y Yolanda Díaz:", t_test_result_sumar.pvalue)
print("\n")

# PP y Feijoo no son normales
mann_whitney_result_pp = mannwhitneyu(data_pp["Puntuacion"], data_nf["Puntuacion"])
print("Medias de puntuación")
print("PP:", data_pp["Puntuacion"].mean())
print("Feijóo:", data_nf["Puntuacion"].mean())
print("Prueba Mann-Whitney U entre PP y Feijóo:", mann_whitney_result_pp.pvalue)
print("\n")

# VOX y Santiago Abascal no son normales
mann_whitney_result_vox = mannwhitneyu(data_vox["Puntuacion"], data_ab["Puntuacion"])
print("Medias de puntuación")
print("VOX:", data_vox["Puntuacion"].mean())
print("Santiago Abascal:", data_ab["Puntuacion"].mean())
print("Prueba Mann-Whitney U entre VOX y Santiago Abascal:", mann_whitney_result_vox.pvalue)
print("\n")
```
